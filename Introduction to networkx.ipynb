```python
g = net.read_edgelist("/content/1684.edges",create_using=net.DiGraph(), nodetype = int)
```

```python
g.number_of_edges() #for getting no of edges 
```

```python
u = g.to_undirected() #for directed graph we can't apply no of connected compnents but we can apply strongly connected so for connected components first made to undirected graph 
net.number_connected_components(u) #used networkx inbuilt function for number of components.
```
